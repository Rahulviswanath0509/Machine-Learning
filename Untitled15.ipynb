```python
import pandas as pd
df=pd.read_csv("titanic.csv")
df.head()
```

```python
new=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
new
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Sex=LabelEncoder()
```

```python
new['Sex_n']=le_Sex.fit_transform(new['Sex'])
```

```python
new
```

```python
new_n=new.drop(['Sex'],axis='columns')
new_n
```

```python
import math
median_newn=math.floor(new_n.Age.median())
median_newn
```

```python
new_n.Age=new_n.Age.fillna(median_newn)
new_n.Age
```

```python
new_n
```

```python
new_n1=new_n.drop(['Survived'],axis='columns')
target=new_n['Survived']
```

```python
target
```

```python
new_n1
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(new_n1,target)
```

```python
model.score(new_n1,target)
```

```python
model.predict([[3,22.0,7.2500,1]])
```
